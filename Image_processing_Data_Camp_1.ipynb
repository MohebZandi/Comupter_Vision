```python
from skimage import data
import numpy as np
```

```python
rocket_image = data.rocket()
coffee_image = data.coffee()
coins_image = data.coins()
```

```python
def show_image(image, title='Image', cmap_type='gray'):
    import matplotlib.pyplot as plt
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
show_image(rocket_image, title='Rocket')
```

```python
show_image(coffee_image, title='Cup of Coffee')
```

```python
show_image(coins_image, title='Coins')
```

```python
# Import the modules from skimage
from skimage import data, color

# Load the rocket image
rocket = data.rocket()

# Convert the image to grayscale
gray_scaled_rocket = color.rgb2gray(rocket)

# Show the original image
show_image(rocket, 'Original RGB image')

# Show the grayscale image
show_image(gray_scaled_rocket, 'Grayscale image')
```

```python
# Load the rocket image
image1 = data.astronaut()
show_image(image1)
```

```python
# Obtaining the red values of the image
red = image1[:, :, 0]
# Obtaining the green values of the image
green = image1[:, :, 1]
# Obtaining the blue values of the image
blue = image1[:, :, 2]
```

```python
plt.imshow(red)
plt.title('Red')
plt.axis('off')
plt.show()
```

```python
plt.imshow(green)
plt.title('green')
plt.axis('off')
plt.show()
```

```python
plt.imshow(blue)
plt.title('blue')
plt.axis('off')
plt.show()
```

```python
plt.imshow(red, cmap="gray")
plt.title('Red')
plt.axis('off')
plt.show()
```

```python
plt.imshow(green, cmap="gray")
plt.title('green')
plt.axis('off')
plt.show()
```

```python
plt.imshow(blue, cmap="gray")
plt.title('blue')
plt.axis('off')
plt.show()
```

```python
# Accessing the shape of the image
image1.shape
```

```python
# Flip the image in up direction
vertically_flipped = np.flipud(image1)
show_image(vertically_flipped, 'Vertically flipped image')
```

```python
# Flip the image in left direction
horizontally_flipped = np.fliplr(image1)
show_image(horizontally_flipped, 'Horizontally flipped image')
```
